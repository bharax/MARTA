## 2. `torch.autograd` package

<font size="3"> The training of a (deep) neural network is not more than performing an optimization on a huge space (=all the NN learnable parameters), based on the **gradient descent**. During training, the learnable parameters are adjusted according to the minimisation of a given loss function $\mathcal{L}$. This loss function must therefore be derived w.r.t. each network parameter. 

<img src="Images/MLPOneLayer.png" width=400 />

To better understand backpropagation, it is first useful to develop an intuition about the relationship between the actual output of a neuron and the correct output for a specific learning example. Let us considere a regression problem using a simple neural network with two inputs $x_1$ and $x_2$ and one single output $y$ (see figure above). For a regression problem, there is no activation layer in the output layer and the loss function is typically a mean square error (MSE) defined as follows, 
$$ \mathcal{L} = \dfrac{1}{N} \sum_{i=1}^N (\hat{y}_i - y_i)^2\, $$
where $\hat{y}_i$ is the network prediction, $y_i$ is the ground truth and $N$ is the number of training data. For a particular training example, the output of a neuron is a weighted sum of all its inputs:
$$\hat{y} = w_1\,x_1 + w_2\,x_2\,.$$
The error depends on the incoming weights to the neuron. This regression problem is discussed below and the loss landscape is also shown to observe the path taken during training by the two parameters ($w_1$ and $w_2$).</font>


```python
# ---
# Loss function
# ---
def MSE(x1,x2,y,w1,w2):
    return np.mean((w1*x1 + w2*x2 - y)**2)

# ---
# Derivative of the loss w.r.t w1 and w2
# ---
def dMSE(x1,x2,y,w1,w2):
    dldw1 = np.mean(2*x1*(w1*x1+w2*x2-y))
    dldw2 = np.mean(2*x2*(w1*x1+w2*x2-y))
    return dldw1, dldw2

# ---
# Training data
# ---
x1      = np.random.rand(100)
x2      = np.random.rand(100)

# ---
# Parameters to fit
# ---
alpha   = 10
beta    = 50

# ---
# Small additional error
# ---
epsilon = np.random.rand(100)*0.005

# ---
# Outputs
# ---
y       = x1*alpha + x2*beta + epsilon

# ---
# Training though gradient descent
# ---
nbEpochs = 10000
w1       = np.zeros(nbEpochs+1)
w1[0]    = np.random.rand(1)
w2       = np.zeros(nbEpochs+1)
w2[0]    = np.random.rand(1)
loss     = np.zeros(nbEpochs)
step     = 0.01

myTimer = time.time()
print("Beginning of the training ...")
for epoch in range(nbEpochs):
    # loss computation
    loss[epoch]  = MSE(x1,x2,y,w1[epoch],w2[epoch])
    # gradient of the loss
    dldw1, dldw2 = dMSE(x1,x2,y,w1[epoch],w2[epoch])
    # update of the learnable parameters
    w1[epoch+1]  = w1[epoch] - step*dldw1
    w2[epoch+1]  = w2[epoch] - step*dldw2

print("... end of the training after %.3f sec" %(time.time() - myTimer))
```

```python
def plot_loss_weights_evolution(nbEpochs,loss,w1,w2,alpha,beta):
    fig, ax = plt.subplots(1,2, constrained_layout=True)
    epoch_vec = np.arange(0,nbEpochs)+1

    # evolution of the loss in a log-log plot
    ax[0].loglog(epoch_vec, loss, 'k-', linewidth=2)
    ax[0].set_ylabel("Loss", fontsize=25)
    ax[0].set_xlabel("Epoch", fontsize=25)
    ax[0].grid(True, which='both')
    ax[0].set_title("Evolution of the loss through epochs", fontsize=20)

    # evolution of the two parameters (=weights)
    ax[1].semilogx(epoch_vec, w1[1:], 'b-', linewidth=2, label="$w_1$")
    ax[1].semilogx(epoch_vec, alpha*np.ones(epoch_vec.shape), 'b--', linewidth=2, label="$\\alpha$")

    ax[1].semilogx(epoch_vec, w2[1:], 'r-', linewidth=2, label="$w_2$")
    ax[1].semilogx(epoch_vec, beta*np.ones(epoch_vec.shape), 'r--', linewidth=2, label="$\\beta$")
    ax[1].set_title("Evolution of the learnable parameters", fontsize=20)

    ax[1].legend(loc="best", fontsize=16, ncol=2)
    ax[1].set_ylabel("Weights", fontsize=25)
    ax[1].set_xlabel("Epoch", fontsize=25)
    ax[1].grid(True, which='both')

    for i in range(2):
        for tick in ax[i].get_xticklabels():
            tick.set_fontsize(15)

        for tick in ax[i].get_yticklabels():
            tick.set_fontsize(15)

    fig.set_size_inches(14,4)
```

```python
plot_loss_weights_evolution(nbEpochs,loss,w1,w2,alpha,beta)
plt.show()
```

```python
# Landscape of the loss
nw1    = 500; nw2    = 500
W1, W2 = np.meshgrid(np.linspace(-50,60,nw1), np.linspace(-50,60,nw2))
LOSS   = np.zeros(W1.shape)
for i in range(nw1):
    for j in range(nw2):
        LOSS[i,j] = MSE(x1,x2,y,W1[i,j],W2[i,j])

fig, ax = plt.subplots(1,1, constrained_layout=True)
im = ax.contourf(W1, W2, LOSS, levels=40, cmap='plasma')

# Evolution of the trajectory
ax.plot(w1,w2,'w-')
ax.plot(w1[0],w2[0],'wx')

ax.set_xlabel("$w_1$", fontsize=25)
ax.set_ylabel("$w_2$", fontsize=25)

ax.set_title("Loss landscape", fontsize=20)

caption_text  = 'Figure 1. The white line represents the\ntrajectory of the two parameters w1 and w2.\n'
caption_text += 'The white cross symbolies the starting point\n(i.e., the initialization of w1 and w2).\n'
caption_text += 'The contour represents the loss valley.\nWe indeed recover a two-dimensional\nquadratic '
caption_text += 'function.'
ax.text(80, -10, caption_text, 
        style='italic', bbox={'facecolor':'white', 'alpha':0, 'pad':10}, fontsize=20)

for tick in ax.get_xticklabels():
    tick.set_fontsize(15)

for tick in ax.get_yticklabels():
    tick.set_fontsize(15)

divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)

plt.colorbar(im, cax=cax)

fig.set_size_inches(15,6)
plt.show()
```

<font size="3">Deriving the full gradient is easy for the simple cases but for classical deep neural networks containing billion of parameters, it would be a nightmare. This is where **backpropagation** comes in. 
    
Essentially, backpropagation is an algorithm to train neural networks based on the chain rule, which makes it more efficient than any naive direct computation of the gradient w.r.t each individual weight. To compute those gradients, `PyTorch` uses automatic differentiation through the built-in differentiation engine called `torch.autograd`. 
    
The `autograd` package creates a computational graph that can be later used to compute derivatives of the output quantities w.r.t the input and other intermediate computations steps. If you want to go into a gentle introduction to `torch.autograd` by PyTorch, we encourage you to read [this](https://pytorch.org/tutorials/beginner/blitz/autograd_tutorial.html). For the time being, let us see how `torch.autograd` can be used to derive an analytical function.</font>

```python
# ---
# Function: x^3 + 10cos(5x) - 10x^2
# ---
def f(x):
    return x**3 + 10*torch.cos(5*x) - 10*x**2

# ---
# Derivative: 3x^2 - 50sin(5x) - 20x
# ---
def df(x):
    return 3*x**2 - 50*torch.sin(5*x) -20*x

print("INFO: .detach() is used to remove the computational graph and .numpy() to transfer to a numpy array.")

fig, ax = plt.subplots(1,2, constrained_layout=True)
x = torch.linspace(-2,10,500)
ax[0].plot(x.detach().numpy(), f(x).detach().numpy(), 'k-', linewidth=2)

df_autograd = torch.zeros(x.shape)
for i in range(x.shape[0]):
    xx = x[i].clone().detach().requires_grad_(True)
    df_autograd[i] = torch.autograd.grad(f(xx), xx)[0].item()

ax[1].plot(x.detach().numpy(), df(x).detach().numpy(), 'k-', linewidth=2, label='Hand-derivation')
ax[1].plot(x.detach().numpy(), df_autograd.detach().numpy(), 'b-', linewidth=2, label='torch.autograd')

ax[0].set_ylabel("$f(x)$", fontsize=25)
ax[1].set_ylabel("$df(x)$", fontsize=25)
ax[1].legend(loc="best", fontsize=15, ncol=1)

ax[0].set_title("$x^3 + 10\\cos(5x) - 10x^2$", fontsize=20)
ax[1].set_title("$3x^2 - 50\\sin(5x) - 20x$", fontsize=20)

for i in range(2):
    ax[i].grid(True)
    ax[i].set_xlabel("$x$", fontsize=25)

    for tick in ax[i].get_xticklabels():
        tick.set_fontsize(15)

    for tick in ax[i].get_yticklabels():
        tick.set_fontsize(15)
    
fig.set_size_inches(12,4)
plt.show()
```

<font size="3">Now we can reuse `torch.autograd` to optimize the previous regression problem. Instead of simply using `torch.autograd.grad` on the MSE, we can directly go one step further by using the function `.backward()` without forgetting to reset the values of the gradients to zero with the `.zero_()` function. </font>

```python
def parametric_function(x1,x2,w1,w2):
    return x1*w1 + x2*w2

# ---
# Transfer to torch.Tensor
# ---
x1_torch = torch.from_numpy(x1)
x2_torch = torch.from_numpy(x2)
y_torch  = torch.from_numpy(y)
param    = torch.randn(2,requires_grad=True)
# ---
# Training though gradient descent
# ---
nbEpochs      = 10000
w1_auto       = np.zeros(nbEpochs+1)
w2_auto       = np.zeros(nbEpochs+1)
loss_auto     = np.zeros(nbEpochs)
lr            = 0.01

myTimer = time.time()
print("Beginning of the training ...")
for epoch in range(nbEpochs):
    y_pred           = parametric_function(x1_torch,x2_torch,param[0],param[1])
    mse              = ((y_torch-y_pred)**2).mean()
    loss_auto[epoch] = mse.item()
    mse.backward()
    param.data      -= lr*param.grad.data
    param.grad.data.zero_()
    w1[epoch+1]  = param.data[0].item()
    w2[epoch+1]  = param.data[1].item()

print("... end of the training after %.3f sec" %(time.time() - myTimer))
```

```python
plot_loss_weights_evolution(nbEpochs,loss,w1,w2,alpha,beta)
plt.show()
```

<font size="3">We end up with the same results but no derivative of the loss has been written down to train this regression problem. The gradients are computed through `mse.backward()` and the parameters are adjusted accordingly with a given learning rate `lr`. Here, the learning rate is kept constant throughout the learning process. However, there are procedures to adapt it dynamically during learning. </font>

<font size="3">The sigmoid activation function (and other bounded activation functions) is prone to the vanishing gradient problem. This issue is well known in the deep learning community. At some point during the training, the gradient reaches almost zero value when going from output to input layer through the backpropagation.

If we consider a simple MLP composed of three layers with $x,w_1,w_2,w_3 \in \mathbb{R}$, such that, 
    $$ f(x; w_1, w_2, w_3) = \sigma( w_3\sigma( w_2\sigma(w_1x) ) ) $$
This expression can be unwrapped as following:
* $a_0 = w_1x $
* $ a_1 = \sigma(a_0) $
* $ a_2 = w_2a_1 $
* $ a_3 = \sigma(a_2) $
* $ a_4 = w_3a_3 $
* $ \hat{y} = \sigma(a_4) $
 
The derviative is then computed using the chain rule to get,
    $$ \dfrac{\partial \hat{y}}{\partial w_1} = 
    \dfrac{\partial \hat{y} }{\partial a_4}
    \dfrac{\partial a_4}{\partial a_3}
    \dfrac{\partial a_3}{\partial a_2}
    \dfrac{\partial a_2}{\partial a_1}
    \dfrac{\partial a_1}{\partial w} = 
    \dfrac{\partial \sigma(a_4)}{\partial a_4}
    w_3
    \dfrac{\partial \sigma(a_2)}{\partial a_2}
    w_2
    \dfrac{\partial \sigma(a_0)}{\partial a_0} x$$

The maximum value of the first derivative of the sigmoid function is $0.25$. Assuming that the weights $w_1,w_2,w_3$ are initiallized randomly from a Gaussian distribution $w_i\sim \mathcal{N}(0,1)$, then we get that, 
    $$ \dfrac{\partial \hat{y}}{\partial w_1} = 
    \underbrace{\dfrac{\partial \sigma(a_4)}{\partial a_4}}_{\leq 0.25}
    \underbrace{w_3}_{\leq 1}
    \underbrace{\dfrac{\partial \sigma(a_2)}{\partial a_2}}_{\leq 0.25}
    \underbrace{w_2}_{\leq 1}
    \underbrace{\dfrac{\partial \sigma(a_0)}{\partial a_0}}_{\leq 0.25} x$$
the derivative shrink to exponentially to zero as we increase the number of layers. This issue may be avoid by using instead unbounded activation function such as ReLU. 
</font>

```python
# ---
# Derivative of the sigmoid
# --
dx   = []
#ddx  = []
X    = np.linspace(-10,10,500)
for item in X:
    x  = torch.tensor(item, requires_grad = True)
    z  = torch.sigmoid(x)
    z.backward(retain_graph=True)
    dx.append(x.grad.data.item())
    #z.backward() 
    #ddx.append(x.grad.data.item())

fig, ax = plt.subplots(1,1, constrained_layout=True)
dx = np.array(dx)
label = ['Layer 1', 'Layer 2', 'Layer 3']
for i, dev in enumerate([dx, dx*dx, dx*dx*dx]):
    ax.plot(X,dev, '-', linewidth=2, label=label[i])

ax.grid(True)
ax.legend(loc='best', fontsize=15)
for tick in ax.get_xticklabels():
    tick.set_fontsize(15)
for tick in ax.get_yticklabels():
    tick.set_fontsize(15)
plt.show()
```

### 4.2. `nn.Module` and `nn.Sequential`

<font size="3">The class `nn.Module` is used to build complex neural networks. Any new networks (or sub-class) which inherit from `nn.Module`, will automatically keep track of the parameters of their components (or properties). To define such a sub-class, it is required to implement the `forward()` method and the constructor `__init__()`.</font>

```python
class MyLinearModel(nn.Module): 
    def __init__(self, in_features, out_features):
        # --- inheritance
        super(MyLinearModel, self).__init__()
        
        # --- parameters
        self.in_features  = in_features
        self.out_features = out_features
        
        # --- linear layer
        self.w = nn.Linear(in_features=self.in_features, 
                           out_features=self.out_features, 
                           bias=True)
    # --- forward pass
    def forward(self, x):
        return self.w(x)
```

```python
# instantiate an object of MyLinearModel with its two inputs arguments
linearModel = MyLinearModel(4,2)

# nn.Module offers useful instructions such as user-friendly printing 
# of the module which summarize the modules contained in it: 
print(linearModel)

# look at the parameters themselves
for param in linearModel.parameters():
    print(param)
```

<font size="3">Now if we want to construct a more complex neural network, we will use `nn.Sequential` that automatically chains modules with each other. Let us construct a multi-layer perceptron (MLP) containing $L$ layers composed of $h_l$ neurons in the layer $l$ with an arbitrary input and output size (`in_size` and `out_size`). </font>

```python
class MySequentialMLP(nn.Module):
    def __init__(self, in_size, hidden_units, out_size, activation="relu"):
        super(MySequentialMLP, self).__init__()
        
        self.hidden_units = hidden_units
        modules           = []
        
        # --- input layer
        lin_in = nn.Linear(in_size, hidden_units[0])
        nn.init.xavier_uniform_(lin_in.weight,gain=nn.init.calculate_gain(activation))
        modules.append(lin_in)
        modules.append(nn.ReLU())
        
        # --- hidden layers (and initialization)
        for i in range(len(hidden_units)-1):
            lin_hidden = nn.Linear(hidden_units[i], hidden_units[i+1])
            nn.init.xavier_uniform_(lin_hidden.weight,gain=nn.init.calculate_gain(activation))
            modules.append(lin_hidden)
            modules.append(nn.ReLU())
        
        # --- output layer
        lin_out = nn.Linear(hidden_units[-1], out_size)
        nn.init.xavier_uniform_(lin_out.weight,gain=nn.init.calculate_gain(activation))
        modules.append(lin_out)
        
        self.net = nn.Sequential(*modules)
        
    # --- forward pass
    def forward(self, x):
        out = self.net(x)
        return out
```

```python
# ---
# Get the number of parameters of each trained neural network
# ---
import functools
import operator

# ---
# https://gist.github.com/ihoromi4/aa16085532358f9fc7937941526d827c
# ---
def get_n_params(model: nn.Module) -> int:
    return sum((functools.reduce(operator.mul, p.size()) for p in model.parameters()))
```

```python
in_size      = 1
out_size     = 1
hidden_units = [15,20,15,10]
mlp          = MySequentialMLP(in_size,hidden_units,out_size)
print("MLP:", mlp)
print("Number of learnable parameters: %d" %(get_n_params(mlp)))
```

```python
# batch of inputs are usually given to train a neural network, because the network takes a scalar entry, 
# we need to reshape the tensor to get torch.Size([40, 1]) compatible with the training.
x = torch.arange(-2, 2, .1).unsqueeze(1) 

# the forward pass is performed and to plot the output, we need to use detach() to convert the tensor
# as numpy matrix (which is implicitely made when you plot a tensor).
y = mlp(x).detach() 

fig, ax = plt.subplots(1,1, constrained_layout=True)
ax.plot(x, y, 'k.-', linewidth=2)
ax.set_xlabel("x", fontsize=25)
ax.set_ylabel("y", fontsize=25)
ax.grid(True)
for tick in ax.get_xticklabels():
    tick.set_fontsize(15)
for tick in ax.get_yticklabels():
    tick.set_fontsize(15)
```

## 4. Optimization with `torch.optim`

<font size="3">We have learned how to construct a neural network using a combination of linear layers and activation functions. We have also learned how to optimize a very simple network using gradient descent and the `torch.autograd` package. As explained, the computation of the gradient is needed to update the parameters of the neural network and to converge through an optimal solution (not necessarily a global minimum). This update procedure is performed by [`torch.optim`](https://pytorch.org/docs/stable/optim.html)). 

According to the documentation of PyTorch: `torch.optim` *is a package implementing various optimization algorithms. Most commonly used methods are already supported, and the interface is general enough so that more sophisticated ones can be also easily integrated in the future.*

Let us consider an example to illustrate the behavior of the stochastic gradient descent (SGD) for the fitting of a quadratic function $$y: = f(x) = x^2 - x + 2$$ 


In the case of gradient descent, ALL the samples in your training are used to update the parameters. Hence, there is a single update of the parameters in a particular iteration. On the other hand, in the case of [stochastic gradient descent](https://pytorch.org/docs/stable/generated/torch.optim.SGD.html), only ONE sample or a SUBSET of samples (randomly selected) is used to update the parameters. This is somehow the extreme case. In this particular case, there are multiple updates of the learnable parameters in a particular iteration. Here is a schematic view of the difference between the gradient descend and the stochastic gradient descent.

<td> <img src="Images/StochasticGradientDescent.png" alt="Drawing" style="width: 300;"/> </td>
</tr></table>

</font>

```python
# unknown function
def fun(x):
    return x**2 - x + 2

# creation of an instance of the SGD class
sgd_optimizer = optim.SGD(params=mlp.parameters(), lr=.001)

# learing steps
nbEpochs = 5000

loss = np.zeros(nbEpochs)

myTimer = time.time()
print("Beginning of the training ...")
for i in range(nbEpochs):
    # randomly generate x values
    x = torch.randn(100, 1)
    
    # prediction
    y_pred = mlp(x)
    
    #  set all the grad values of the parameters of our net to zero
    sgd_optimizer.zero_grad()
    
    # computation of the loss and its gradient
    loss_item = ((fun(x) - y_pred)**2).mean()
    loss_item.backward()
    
    # always monitor your loss !
    loss[i] = loss_item.item()
    
    # update of the parameters 
    sgd_optimizer.step()
    
print("... end of the training after %.3f sec" %(time.time() - myTimer))
```

```python
fig, ax = plt.subplots(1,2, constrained_layout=True)

ax[0].loglog(np.arange(1,nbEpochs+1), loss, 'k.-', linewidth=2)
ax[0].set_xlabel("Epoch", fontsize=25)
ax[0].set_ylabel("Loss", fontsize=25)
    
x = torch.arange(-2, 2, .1).unsqueeze(1) 
y = mlp(x).detach() 

ax[1].plot(x, fun(x), 'k-', linewidth=2, label='$f(x)$')
ax[1].plot(x, y, 'r.-', linewidth=2, label='$\hat{y}$')
ax[1].legend(loc="upper right", fontsize=15)

ax[1].set_xlabel("x", fontsize=25)
ax[1].set_ylabel("y", fontsize=25)

for i in range(2):
    ax[i].grid(True)
    for tick in ax[i].get_xticklabels():
        tick.set_fontsize(15)
    for tick in ax[i].get_yticklabels():
        tick.set_fontsize(15)
        
fig.set_size_inches(14,4)
```

## 5. Database and DataLoader

### 5.1 Under-fitting and Over-fitting

<font size="3">While training a (deep) neural network, you need to have three distinct sets: 
1. a **training** set,
2. a **validation** set and,
3. a **test** set.

The training and validation sets are used during the optimization of the network. The **training set** contains the data that is used to fit the learnable parameters while the **validation set** contains data that is used to evaluate the performance of the current parameters. We hence obtain a training loss and a validation loss. In good training, the training and validation losses should be close to each other and should decrease with the number of epochs. However, at some point, the validation loss will start to increase. The training should be stopped at that point, this is called _early stop_. Above this point, the neural network overfits the data (also called the _overfitting_ regime). In this regime, the model starts to be very close to the training data but does not generalize well. This behavior is to be avoided otherwise the network won't perform correctly in production.
    
The question is: What happens if the hypothesis space $\mathcal{F}$ in which candidate function $f$ are either too "_simple_" or too "_complex_" w.r.t. the true function we want to model (i.e., the true data generating process)? Do we underfit or overfit our problem? This issue can be also related to the [bias-variance trade-off](https://en.wikipedia.org/wiki/Bias%E2%80%93variance_tradeoff). Let us illustrate the issue with a simple cubic function defined as follows:
    $$y = f(x) + \epsilon = \alpha x^3\,\, \mbox{ where }\,\, \epsilon \sim N(0,\sigma_e^2) $$

 </font>

```python
def Ffun(x, coef=[-1,-1,15,-10]):
    return coef[0]*x**3 + coef[1]*x**2 + coef[2]*x + coef[3]

fig, ax = plt.subplots(1,1, constrained_layout=True)
x       = np.arange(-5,5,0.05)
x_train = np.arange(-5,5,0.5)
epsilon = np.random.normal(0,5,x_train.shape[0])
y       = Ffun(x_train)+epsilon
ax.plot(x_train, y, 'bs', alpha=1, label="$f(x) + \\epsilon$")
ax.plot(x, Ffun(x), 'k-', linewidth=4, label="$f(x)$")

# ---
# Polyfit
# ---
p1 = np.polyfit(x_train, y, 1)
R1 = np.mean((y-np.polyval(p1,x_train))**2)
ax.plot(x, np.polyval(p1,x), 'g-', linewidth=3, label="$\\hat{y}$ at order 1 with $\\hat{R}(f,d)=$ %.2f"%(R1))

p2 = np.polyfit(x_train, y, 2)
R2 = np.mean((y-np.polyval(p2,x_train))**2)
ax.plot(x, np.polyval(p2,x), 'r-', linewidth=3, label="$\\hat{y}$ at order 2 with $\\hat{R}(f,d)=$ %.2f"%(R2))

p3 = np.polyfit(x_train, y, 3)
R3 = np.mean((y-np.polyval(p3,x_train))**2)
ax.plot(x, np.polyval(p3,x), 'm-', linewidth=3, label="$\\hat{y}$ at order 3 with $\\hat{R}(f,d)=$ %.2f"%(R3))

p10 = np.polyfit(x_train, y, 10)
R10 = np.mean((y-np.polyval(p10,x_train))**2)
ax.plot(x, np.polyval(p10,x), '-', linewidth=3, label="$\\hat{y}$ at order 10 with $\\hat{R}(f,d)=$ %.2f"%(R10))


ax.grid(True)
for tick in ax.get_xticklabels():
    tick.set_fontsize(15)
for tick in ax.get_yticklabels():
    tick.set_fontsize(15)

ax.set_xlabel("$x$", fontsize=15)
ax.legend(loc='best', fontsize=15)
ax.set_xlim(-5,5)
ax.set_ylim(-80,20)

fig.set_size_inches(10,8)

plt.show()
```

<font size="3">$\hat{R}(f,d)$ is an estimate of the empirical risk (or training error) and it is defined as
$$ \hat{R}(f,d) = \dfrac{1}{N} \sum_{(\mathbf{x}_i,y_i) \in \mathbf{d}} l(y_i, f(\mathbf{x}_i))\,, $$
where $N$ is the size of the training data. Most ML algorithms implement this empirical risk minimization. See how this risk is minimized as the polynomial degree is increased. However, we can clearly see that the blue curve does not generalize on the true data (i.e., the black curve). This simple example is a good observation of the underfitting/overfitting issue. The capacity of a hypothesis space is related to the ability to find a good model, regardless of its complexity. The capacity is controlled by a set of hyper-parameters of the learning algorithm. For a polynomial fitting, the capacity of the model is directly linked to the interpolant degree. For a neural network, these hyper-parameters can be: the number of layers, the number of training iterations, regularization terms, ...

Hence, to avoid this issue, an estimate of the risk $\hat{R}(f^d_{\star},\mathbf{d}_{\mbox{test}})$ can be computed on a distinct set, called the **test set** to evaluate the performance of the model. It should not be used for model selection. 
</font>

### `DataLoader`

<font size="3">There exists online databases that you can download and use as it is. However, most of the time, you have our own data and you want to make them accessible for the training. `PyTorch` offers the `DataLoader` class. For a complete review of `DataLoader`, we encourage you to read [this](https://pytorch.org/docs/stable/data.html). To summarize quicky, it exists two type of datasets: (a) the map-style datasets, and (b) iterable-style datasets. According to the documentation, 

* *A map-style dataset is one that implements the `__getitem__()` and `__len__()` protocols, and represents a map from (possibly non-integral) indices/keys to data samples.*

* *An iterable-style dataset is an instance of a subclass of IterableDataset that implements the `__iter__()` protocol, and represents an iterable over data samples. This type of datasets is particularly suitable for cases where random reads are expensive or even improbable, and where the batch size depends on the fetched data.*

Let us construct an iteratble-style dataset for the fitting of our quadratique function. The library `sklearn` also offer the opportunity to create database and to split them into training and validation database, as well as splitting according to K-fold cross-validation. </font>

```python
class MyIterableDataset(torch.utils.data.IterableDataset):
    def __init__(self, x, perc, epsilon, train=True):
        super(MyIterableDataset).__init__()
        
        self.nbData = x.shape[0]
        self.x      = x
        self.y      = fun(self.x).view((self.nbData,1)) 
        self.y     += torch.from_numpy(epsilon).view((self.nbData,1)).to(torch.float32)
        
        self.train  = train
        self.test   = not(train)
        self.size   = int(self.nbData*(1-perc))*self.test + int(self.nbData*perc)*self.train
        self.offset = int(self.nbData*(1-perc))*self.train
        
    def __iter__(self):
        worker_info = torch.utils.data.get_worker_info()
        if worker_info is None:  # single-process data loading, return the full iterator
            iter_start = self.offset
            iter_end   = self.offset + self.size
        else:  
            print("not implemented, read the documentation")
        index = torch.arange(iter_start,iter_end)
        buf   = torch.cat((self.x[index,:], self.y[index,:]), 1)
        return iter(buf)
```

```python
nbData = 2000
x      = torch.linspace(-2,2,nbData).view((nbData,1))
rand_indx = torch.randperm(len(x))

epsilon     = np.random.normal(0,0.5,x.shape[0])
trainData   = MyIterableDataset(x[rand_indx], 0.9, epsilon, train=True)
trainLoader = torch.utils.data.DataLoader(trainData, batch_size=256, num_workers=0, pin_memory=False)

testData   = MyIterableDataset(x[rand_indx], 0.9, epsilon, train=False)
testLoader = torch.utils.data.DataLoader(testData, batch_size=256, num_workers=0)
```

```python
# computation of the test loss
def test_loss(testLoader):
    test_loss = 0
    nb        = 0
    for data in testLoader:
        batch_sz = data.shape[0]
        x        = data[:,0].view((batch_sz,1))
        y        = data[:,1].view((batch_sz,1))
        y_pred   = mlp(x)        
        batch_loss = ((y - y_pred)**2).mean()
        test_loss += batch_loss.detach()
        nb         += 1
        
    return test_loss/nb

# model
mlp = MySequentialMLP(in_size,hidden_units,out_size)

# creation of an instance of the SGD class
sgd_optimizer = optim.SGD(params=mlp.parameters(), lr=.001)

# learing steps
nbEpochs = 5000

loss     = np.zeros(nbEpochs)
lossTest = np.zeros(nbEpochs)

myTimer = time.time()
print("Beginning of the training ...")
for i in range(nbEpochs):
    mlp.train()

    train_loss = 0
    nb         = 0
    for data in trainLoader:
        batch_sz = data.shape[0]
        x        = data[:,0].view((batch_sz,1))
        y        = data[:,1].view((batch_sz,1))
        y_pred   = mlp(x)

        #  set all the grad values of the parameters of our net to zero
        sgd_optimizer.zero_grad()        
        
        # computation of the loss and its gradient
        batch_loss = ((y - y_pred)**2).mean()
        batch_loss.backward()
    
        # update of the parameters 
        sgd_optimizer.step()
        
        # save loss
        train_loss += batch_loss.detach()
        nb         += 1
    
    # always record your loss
    loss[i]     = train_loss/nb
    lossTest[i] = test_loss(testLoader)
    if(i%10==0):
        print("Epoch: %6d, train_loss=%.6e, val_loss=%.6e" %(i,loss[i],lossTest[i]))
print("... end of the training after %.3f sec" %(time.time() - myTimer))
```

```python
fig, ax = plt.subplots(1,2, constrained_layout=True)

ax[0].loglog(np.arange(1,nbEpochs+1), loss, 'k.-', linewidth=2, label="Train Loss")
ax[0].loglog(np.arange(1,nbEpochs+1), lossTest, 'b.-', linewidth=2, label="Val. Loss")
ax[0].set_xlabel("Epoch", fontsize=25)
ax[0].set_ylabel("Loss", fontsize=25)
ax[0].legend(loc="upper right", fontsize=15)
ax[0].set_ylim(1.8e-1,3e-1)

x  = torch.arange(-2, 2, .1).unsqueeze(1)  
y  = mlp(x).detach() 

ax[1].plot(x, fun(x), 'k-', linewidth=2, label='$f(x)$')
ax[1].plot(x, y, 'r.-', linewidth=2, label='$\hat{y}$')
ax[1].legend(loc="upper right", fontsize=15)
ax[1].set_xlabel("x", fontsize=25)
ax[1].set_ylabel("y", fontsize=25)

for i in range(2):
    ax[i].grid(True)
    for tick in ax[i].get_xticklabels():
        tick.set_fontsize(15)
    for tick in ax[i].get_yticklabels():
        tick.set_fontsize(15)
        
fig.set_size_inches(10,4)
```

<font size="3">The loss is better, it does not present crazy oscillations as previously. We observe also the good fit of the network w.r.t. the analytical expression of the quadratic function $f(x)$. If you observe correctly the red curve, you will see that it appears as three *straight lines*. It is due to the activation function used in this training. The ReLU is a piecewise linear function and the prediction also appears as a *piecewise linear function* (roughly speaking). To enhance the smoothness of the prediction, you can add more hidden layers into the model. It will increase the complexity of the model by increasing the number of parameters. </font>
