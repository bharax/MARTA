### 2.1. `nn.Module` and `nn.Sequential`

<font size="3">The class `nn.Module` is used to build complex neural networks. Any new networks (or sub-class) which inherit from `nn.Module` will automatically keep track of the parameters of their components (or properties). To define such a sub-class, it is required to implement the `forward()` method and the constructor `__init__()`.
<img src="Images/nn.png" width=400 />
</font>

<font size="3"><b>Neuron computation:</b>
<img src="Images/Neuron.png" width=500 />
where $w_i$ are the weights and $b$ is a bias. All these parameters are the neural network (NN) learnable parameters that will be fitted during the training. 
</font>

```python
class MyLinearModel(nn.Module): 
    def __init__(self, in_features, out_features):
        # --- inheritance
        super(MyLinearModel, self).__init__()
        
        # --- parameters
        self.in_features  = in_features
        self.out_features = out_features
        
        # --- linear layer
        self.w = nn.Linear(in_features=self.in_features, 
                           out_features=self.out_features, 
                           bias=True)
    # --- forward pass
    def forward(self, x):
        return self.w(x)
```

```python
# instantiate an object of MyLinearModel with its two inputs arguments
linearModel = MyLinearModel(4,2)

# nn.Module offers useful instructions such as user-friendly printing 
# of the module which summarize the modules contained in it: 
print(linearModel)

# look at the parameters themselves
params = []
for param in linearModel.parameters():
    print(param)
    params.append(param)
```

```python
xx_in = torch.ones(4)
# forward pass via module
print("Output:", linearModel(xx_in))
# manual forward pass
weights = params[0]
bias    = params[1]
print("Output:",weights@xx_in+bias)
```

<font size="3">The sigmoid activation function (and other bounded activation functions) is prone to the vanishing gradient problem. This issue is well-known in the deep learning community. At some point during the training, the gradient reaches almost zero value when going from the output to the input layer through the backpropagation. We will treat this gradient notion in the next section.  This issue may be avoided by using instead unbounded activation functions such as ReLU. It exists many other techniques to treat this problem such as skip connection, good weights initialization, data normalization, batch normalization + ReLU, ...). 
</font>

<font size="3">
<div class="alert alert-block alert-warning">
<b>To go further ...</b> If we consider a simple MLP composed of three layers with $x,w_1,w_2,w_3 \in \mathbb{R}$, such that, 
    $$ f(x; w_1, w_2, w_3) = \sigma( w_3\sigma( w_2\sigma(w_1x) ) ) $$
This expression can be unwrapped as follows:
<ul>
    <li>$a_0 = w_1x $</li> 
    <li>$ a_1 = \sigma(a_0) $</li>
    <li>$ a_2 = w_2a_1 $</li>
    <li>$ a_3 = \sigma(a_2) $</li>
    <li>$ a_4 = w_3a_3 $</li>
    <li>$ \hat{y} = \sigma(a_4) $ </li>
</ul>

The derivative is then computed using the chain rule to get,
    $$ \dfrac{\partial \hat{y}}{\partial w_1} = 
    \dfrac{\partial \hat{y} }{\partial a_4}
    \dfrac{\partial a_4}{\partial a_3}
    \dfrac{\partial a_3}{\partial a_2}
    \dfrac{\partial a_2}{\partial a_1}
    \dfrac{\partial a_1}{\partial w} = 
    \dfrac{\partial \sigma(a_4)}{\partial a_4}
    w_3
    \dfrac{\partial \sigma(a_2)}{\partial a_2}
    w_2
    \dfrac{\partial \sigma(a_0)}{\partial a_0} x$$

The maximum value of the first derivative of the sigmoid function is $0.25$. Assuming that the weights $w_1,w_2,w_3$ are initialized randomly from a Gaussian distribution $w_i\sim \mathcal{N}(0,1)$, then we get that, 
    $$ \dfrac{\partial \hat{y}}{\partial w_1} = 
    \underbrace{\dfrac{\partial \sigma(a_4)}{\partial a_4}}_{\leq 0.25}
    \underbrace{w_3}_{\leq 1}
    \underbrace{\dfrac{\partial \sigma(a_2)}{\partial a_2}}_{\leq 0.25}
    \underbrace{w_2}_{\leq 1}
    \underbrace{\dfrac{\partial \sigma(a_0)}{\partial a_0}}_{\leq 0.25} x$$
the derivative shrink to exponentially to zero as we increase the number of layers.
</div>
</font>

```python
# ---
# Derivative of the sigmoid
# --
dx   = []
#ddx  = []
X    = np.linspace(-10,10,500)
for item in X:
    x  = torch.tensor(item, requires_grad = True)
    z  = torch.sigmoid(x)
    z.backward(retain_graph=True)
    dx.append(x.grad.data.item())
    #z.backward() 
    #ddx.append(x.grad.data.item())

fig, ax = plt.subplots(1,1, constrained_layout=True)
dx = np.array(dx)
label = ['Layer 1', 'Layer 2', 'Layer 3']
for i, dev in enumerate([dx, dx*dx, dx*dx*dx]):
    ax.plot(X,dev, '-', linewidth=2, label=label[i])

ax.grid(True)
ax.legend(loc='best', fontsize=15)
for tick in ax.get_xticklabels():
    tick.set_fontsize(15)
for tick in ax.get_yticklabels():
    tick.set_fontsize(15)
plt.show()
```

### 2.3 Putting all together

<font size="3">Now if we want to construct a more complex neural network, we will use `nn.Sequential` that automatically chains modules with each other. Let us construct a multi-layer perceptron (MLP) containing $L$ layers composed of $h_l$ neurons in the layer $l$ with an arbitrary input and output size (`in_size` and `out_size`). 
</font>

```python
class MySequentialMLP(nn.Module):
    def __init__(self, in_size, hidden_units, out_size, activation="relu"):
        super(MySequentialMLP, self).__init__()
        
        self.hidden_units = hidden_units
        modules           = []
        
        # --- input layer
        lin_in = nn.Linear(in_size, hidden_units[0])
        nn.init.xavier_uniform_(lin_in.weight,gain=nn.init.calculate_gain(activation))
        modules.append(lin_in)
        modules.append(nn.ReLU())
        
        # --- hidden layers (and initialization)
        for i in range(len(hidden_units)-1):
            lin_hidden = nn.Linear(hidden_units[i], hidden_units[i+1])
            nn.init.xavier_uniform_(lin_hidden.weight,gain=nn.init.calculate_gain(activation))
            modules.append(lin_hidden)
            modules.append(nn.ReLU())
        
        # --- output layer
        lin_out = nn.Linear(hidden_units[-1], out_size)
        nn.init.xavier_uniform_(lin_out.weight,gain=nn.init.calculate_gain(activation))
        modules.append(lin_out)
        
        self.net = nn.Sequential(*modules)
        
    # --- forward pass
    def forward(self, x):
        out = self.net(x)
        return out
```

```python
in_size      = 1
out_size     = 1
hidden_units = [15,20,15,10]
mlp          = MySequentialMLP(in_size,hidden_units,out_size)
print("MLP:", mlp)
print("Number of learnable parameters: %d" %(get_n_params(mlp)))
```

```python
# batch of inputs are usually given to train a neural network, because the network takes a scalar entry, 
# we need to reshape the tensor to get torch.Size([40, 1]) compatible with the training.
x = torch.arange(-2, 2, .1).unsqueeze(1) 

# the forward pass is performed and to plot the output, we need to use detach() to convert the tensor
# as numpy matrix (which is implicitely made when you plot a tensor).
y = mlp(x).detach() 

fig, ax = plt.subplots(1,1, constrained_layout=True)
ax.plot(x, y, 'k.-', linewidth=2)
ax.set_xlabel("x", fontsize=25)
ax.set_ylabel("y", fontsize=25)
ax.grid(True)
for tick in ax.get_xticklabels():
    tick.set_fontsize(15)
for tick in ax.get_yticklabels():
    tick.set_fontsize(15)
```

## 3. `torch.autograd` package

<font size="3"> The training of a (deep) neural network is not more than performing an optimization on a huge space (=all the NN learnable parameters), based on the **gradient descent**. During training, the learnable parameters are adjusted according to the minimization of a given loss function $\mathcal{L}$. This loss function must therefore be derived w.r.t. each network parameter. 

<img src="Images/MLPOneLayer.png" width=400 />

To better understand backpropagation, it is first useful to develop an intuition about the relationship between the actual output of a neuron and the correct output for a specific learning example. Let us consider a regression problem using a simple neural network with two inputs $x_1$ and $x_2$ and one single output $y$ (see figure above). For a regression problem, there is no activation layer in the output layer and the loss function is typically a mean square error (MSE) defined as follows, 
$$ \mathcal{L} = \dfrac{1}{N} \sum_{i=1}^N (\hat{y}_i - y_i)^2\, $$
where $\hat{y}_i$ is the network prediction, $y_i$ is the ground truth and $N$ is the number of training data. For a particular training example, the output of a neuron is a weighted sum of all its inputs:
$$\hat{y} = w_1\,x_1 + w_2\,x_2\,.$$
The error depends on the incoming weights to the neuron. This regression problem is discussed below and the loss landscape is also shown to observe the path taken during training by the two parameters ($w_1$ and $w_2$).

To optimize this little problem, we need to compute the derivative of $\mathcal{L}$ with respect to the two parameters $w_1$ and $w_2$ to get,
$$ \dfrac{\partial \mathcal{L}}{\partial w_1} = 2 x_1(w_1 x_1 + w_2 x_2) \hspace{5mm} \mbox{ and } \hspace{5mm} \dfrac{\partial \mathcal{L}}{\partial w_2} = 2 x_2(w_1 x_1 + w_2 x_2)$$
Then, the learnable parameters $w_1$ and $w_2$ are updated according to:
$$ w_{1,t+1} = w_{1,t} - \gamma \dfrac{\partial \mathcal{L}}{\partial w_1} \hspace{5mm} \mbox{ and } \hspace{5mm}  w_{2,t+1} = w_{2,t} - \gamma \dfrac{\partial \mathcal{L}}{\partial w_2} $$
where $\gamma$ is the learning rate. We will discuss further the impact of this learning rate on the overall training.
</font>


```python
# ---
# Loss function
# ---
def MSE(x1,x2,y,w1,w2):
    return np.mean((w1*x1 + w2*x2 - y)**2)

# ---
# Derivative of the loss w.r.t w1 and w2
# ---
def dMSE(x1,x2,y,w1,w2):
    dldw1 = np.mean(2*x1*(w1*x1+w2*x2-y))
    dldw2 = np.mean(2*x2*(w1*x1+w2*x2-y))
    return dldw1, dldw2

# ---
# Training data
# ---
x1      = np.random.rand(100)
x2      = np.random.rand(100)

# ---
# Parameters to fit
# ---
alpha   = 10
beta    = 50

# ---
# Small additional error
# ---
epsilon = np.random.rand(100)*0.005

# ---
# Outputs
# ---
y       = x1*alpha + x2*beta + epsilon

# ---
# Training though gradient descent
# ---
nbEpochs = 10000
w1       = np.zeros(nbEpochs+1)
w1[0]    = np.random.rand(1)
w2       = np.zeros(nbEpochs+1)
w2[0]    = np.random.rand(1)
loss     = np.zeros(nbEpochs)
step     = 0.01

myTimer = time.time()
print("Beginning of the training ...")
for epoch in range(nbEpochs):
    # loss computation
    loss[epoch]  = MSE(x1,x2,y,w1[epoch],w2[epoch])
    # gradient of the loss
    dldw1, dldw2 = dMSE(x1,x2,y,w1[epoch],w2[epoch])
    # update of the learnable parameters
    w1[epoch+1]  = w1[epoch] - step*dldw1
    w2[epoch+1]  = w2[epoch] - step*dldw2

print("... end of the training after %.3f sec" %(time.time() - myTimer))
```

```python
def plot_loss_weights_evolution(nbEpochs,loss,w1,w2,alpha,beta):
    fig, ax = plt.subplots(1,2, constrained_layout=True)
    epoch_vec = np.arange(0,nbEpochs)+1

    # evolution of the loss in a log-log plot
    ax[0].loglog(epoch_vec, loss, 'k-', linewidth=2)
    ax[0].set_ylabel("Loss", fontsize=25)
    ax[0].set_xlabel("Epoch", fontsize=25)
    ax[0].grid(True, which='both')
    ax[0].set_title("Evolution of the loss through epochs", fontsize=20)

    # evolution of the two parameters (=weights)
    ax[1].semilogx(epoch_vec, w1[1:], 'b-', linewidth=2, label="$w_1$")
    ax[1].semilogx(epoch_vec, alpha*np.ones(epoch_vec.shape), 'b--', linewidth=2, label="$\\alpha$")

    ax[1].semilogx(epoch_vec, w2[1:], 'r-', linewidth=2, label="$w_2$")
    ax[1].semilogx(epoch_vec, beta*np.ones(epoch_vec.shape), 'r--', linewidth=2, label="$\\beta$")
    ax[1].set_title("Evolution of the learnable parameters", fontsize=20)

    ax[1].legend(loc="best", fontsize=16, ncol=2)
    ax[1].set_ylabel("Weights", fontsize=25)
    ax[1].set_xlabel("Epoch", fontsize=25)
    ax[1].grid(True, which='both')

    for i in range(2):
        for tick in ax[i].get_xticklabels():
            tick.set_fontsize(15)

        for tick in ax[i].get_yticklabels():
            tick.set_fontsize(15)

    fig.set_size_inches(14,4)
```

<font size="3">Deriving the full gradient is easy for simple cases but for classical deep neural networks containing billion of parameters, it would be a nightmare. This is where **backpropagation** comes in. 
    
Essentially, backpropagation is an algorithm that quickly computes the gradient of the loss function w.r.t. each parameter based on the chain rule, which makes it more efficient than any naive direct computation of the gradient w.r.t each individual weight. To compute those gradients, `PyTorch` uses automatic differentiation through the built-in differentiation engine called `torch.autograd`. 
    
The `autograd` package creates a computational graph that can be later used to compute derivatives of the output quantities w.r.t the input and other intermediate computations steps. If you want to go into a gentle introduction to `torch.autograd` by PyTorch, we encourage you to read [this](https://pytorch.org/tutorials/beginner/blitz/autograd_tutorial.html). For the time being, let us see how `torch.autograd` can be used to derive an analytical function.</font>

```python
# ---
# Function: x^3 + 10cos(5x) - 10x^2
# ---
def f(x):
    return x**3 + 10*torch.cos(5*x) - 10*x**2

# ---
# Derivative: 3x^2 - 50sin(5x) - 20x
# ---
def df(x):
    return 3*x**2 - 50*torch.sin(5*x) -20*x

print("INFO: .detach() is used to remove the computational graph and .numpy() to transfer to a numpy array.")

fig, ax = plt.subplots(1,2, constrained_layout=True)
x = torch.linspace(-2,10,500)
ax[0].plot(x.detach().numpy(), f(x).detach().numpy(), 'k-', linewidth=2)

df_autograd = torch.zeros(x.shape)
for i in range(x.shape[0]):
    xx = x[i].clone().detach().requires_grad_(True)
    df_autograd[i] = torch.autograd.grad(f(xx), xx)[0].item()

ax[1].plot(x.detach().numpy(), df(x).detach().numpy(), 'k-', linewidth=2, label='Hand-derivation')
ax[1].plot(x.detach().numpy(), df_autograd.detach().numpy(), 'b-', linewidth=2, label='torch.autograd')

ax[0].set_ylabel("$f(x)$", fontsize=25)
ax[1].set_ylabel("$df(x)$", fontsize=25)
ax[1].legend(loc="best", fontsize=15, ncol=1)

ax[0].set_title("$x^3 + 10\\cos(5x) - 10x^2$", fontsize=20)
ax[1].set_title("$3x^2 - 50\\sin(5x) - 20x$", fontsize=20)

for i in range(2):
    ax[i].grid(True)
    ax[i].set_xlabel("$x$", fontsize=25)

    for tick in ax[i].get_xticklabels():
        tick.set_fontsize(15)

    for tick in ax[i].get_yticklabels():
        tick.set_fontsize(15)
    
fig.set_size_inches(12,4)
plt.show()
```

<font size="3">Now we can reuse `torch.autograd` to automatically and transparently compute the gradient within the previous regression problem. Instead of simply using `torch.autograd.grad` on the MSE, we can directly go one step further by using the function `.backward()` without forgetting to reset the values of the gradients to zero with the `.zero_()` function. </font>

```python
def parametric_function(x1,x2,w1,w2):
    return x1*w1 + x2*w2

# ---
# Transfer to torch.Tensor
# ---
x1_torch = torch.from_numpy(x1)
x2_torch = torch.from_numpy(x2)
y_torch  = torch.from_numpy(y)
param    = torch.randn(2,requires_grad=True)
# ---
# Training though gradient descent
# ---
nbEpochs      = 10000
w1_auto       = np.zeros(nbEpochs+1)
w2_auto       = np.zeros(nbEpochs+1)
loss_auto     = np.zeros(nbEpochs)
lr            = 0.01

myTimer = time.time()
print("Beginning of the training ...")
for epoch in range(nbEpochs):
    y_pred           = parametric_function(x1_torch,x2_torch,param[0],param[1])
    mse              = ((y_torch-y_pred)**2).mean()
    loss_auto[epoch] = mse.item()
    mse.backward()
    param.data      -= lr*param.grad.data
    param.grad.data.zero_()
    w1[epoch+1]  = param.data[0].item()
    w2[epoch+1]  = param.data[1].item()

print("... end of the training after %.3f sec" %(time.time() - myTimer))
```

<font size="3">We end up with the same results but no derivative of the loss has been written down to train this regression problem. The gradients are computed through `mse.backward()` and the parameters are adjusted according to the gradient descent method with a given learning rate `lr` (which is a learning hyperparameter). 
$$ \theta_{t+1} = \theta_t - \texttt{lr}\,g_t$$
where $\theta$ is a parameter, $t$ is an iteration, $\gamma$ the learning rate and $g$ the gradient, computed with backpropagation. Here, the learning rate is kept constant throughout the learning process. However, there are procedures to adapt it dynamically during learning. </font>

## 4. Optimization with `torch.optim`

<font size="3">We have learned how to construct a neural network using a combination of linear layers and activation functions. We have also learned how to optimize a very simple network using gradient descent and the `torch.autograd` package. As explained, the computation of the gradient is needed to update the parameters of the neural network and to converge through an optimal solution (not necessarily a global minimum). This update procedure is performed by [`torch.optim`](https://pytorch.org/docs/stable/optim.html). 

According to the documentation of PyTorch: 

`torch.optim` *is a package implementing various optimization algorithms. Most commonly used methods are already supported, and the interface is general enough so that more sophisticated ones can be also easily integrated in the future.*

The **main optimization procedure** used to train neural network is the Gradient Descent:
<div class="alert alert-block alert-success">
<b>Gradient Descent (GD)</b> ALL the samples in your training are used to update the parameters. Hence, there is a single update of the parameters in a particular iteration. 
</div>

For memory concern and efficiency, one may define **unbiased estimators** of the gradient which leads to iterative methods for optimizing the loss function:
<div class="alert alert-block alert-success">
<b>Stochastic Gradient Descent (SGD)</b> only ONE sample (randomly selected) is used to update the parameters (see this <a href="https://pytorch.org/docs/stable/generated/torch.optim.SGD.html">page</a>). This is somehow the extreme case. In this particular case, there are multiple updates of the learnable parameters in a particular iteration. It is the "<i>drunk version of the GD</i>" (roughly speaking, see figure). Note that this optimizer takes time to compute and becomes inefficient for large databases. Nonetheless, it leads to good generalization properties because the gradient estimate is very noisy, which may help to escape from local minima. 
</div>

<div class="alert alert-block alert-success">
<b>Mini-batching</b> consists of visiting the samples in mini-batches (i.e. a randomly selected subset of the training database) and the parameters are updated after each batch visit. The batch size $B$ helps to control the variance of the gradient estimates. 
</div>

To help for faster convergence, **momentum** (*if the slope is very steep, we will reach the bottom faster*) has been added to the update of the parameters (e.g. Nesterov momentum). Adaptative learning rate methods have also been created such as AdaGrad, RMSProp, Adam.

<font size="3"> Let us consider an example to illustrate the behavior of the stochastic gradient descent (SGD) for the fitting of a quadratic function $$y: = f(x) = x^2 - x + 2$$ </font>

```python
# unknown function
def fun(x,mean=0,std=0.1):
    return x**2 - x + 2 + torch.normal(mean*torch.ones(x.shape[0]),std*torch.ones(x.shape[0])).view(-1,1)

# creation of an instance of the SGD class
sgd_optimizer = optim.SGD(params=mlp.parameters(), lr=.001)

# learing steps
nbEpochs = 5000

# randomly generate x values
x = torch.randn(100, 1)

# initialize loss vector
loss = np.zeros(nbEpochs)

myTimer = time.time()
print("Beginning of the training ...")
for i in range(nbEpochs):
    # prediction
    y_pred = mlp(x)
    
    #  set all the grad values of the parameters of our net to zero
    sgd_optimizer.zero_grad()
    
    # computation of the loss and its gradient
    loss_item = ((fun(x) - y_pred)**2).mean()
    loss_item.backward()
    
    # always monitor your loss !
    loss[i] = loss_item.item()
    
    # update of the parameters 
    sgd_optimizer.step()
    
print("... end of the training after %.3f sec" %(time.time() - myTimer))
```

## 5. Database and DataLoader

### 5.1 Under-fitting and Over-fitting

<font size="3">While training a (deep) neural network, you need to have three **distinct** and **independent** sets: 
1. a **training** set,
2. a **validation** set and,
3. a **test** set.

The training and validation sets are used during the optimization of the network. The **training set** contains the data that is used to fit the learnable parameters while the **validation set** contains data that is used to evaluate the performance of the current parameters. We hence obtain a training loss and a validation loss. In good training, the training and validation losses should be close to each other and should decrease with the number of epochs. However, at some point, the validation loss will start to increase. The training should be stopped at that point, this is called _early stop_. Above this point, the neural network overfits the data (also called the _overfitting_ regime). In this regime, the model starts to be very close to the training data but does not generalize well. This behavior is to be avoided otherwise the network won't perform correctly in production.
    
The question is: What happens if the hypothesis space $\mathcal{F}$ in which candidate function $f$ are either too "_simple_" or too "_complex_" w.r.t. the true function we want to model (i.e., the true data generating process)? Do we underfit or overfit our problem? This issue can be also related to the [bias-variance trade-off](https://en.wikipedia.org/wiki/Bias%E2%80%93variance_tradeoff). Let us illustrate the issue with a simple cubic function defined as follows:
    $$y = f(x) + \epsilon = -x^3 - x^2 + 15x - 10 + \epsilon \,\, \mbox{ where }\,\, \epsilon \sim N(0,\sigma_e^2) $$

 </font>

```python
def Ffun(x, coef=[-1,-1,15,-10]):
    return coef[0]*x**3 + coef[1]*x**2 + coef[2]*x + coef[3]

fig, ax = plt.subplots(1,1, constrained_layout=True)
x       = np.arange(-5,5,0.05)
x_train = np.arange(-5,5,0.5)
epsilon = np.random.normal(0,5,x_train.shape[0])
y       = Ffun(x_train)+epsilon
ax.plot(x_train, y, 'bs', alpha=1, label="$f(x) + \\epsilon$")
ax.plot(x, Ffun(x), 'k-', linewidth=4, label="$f(x)$")

# ---
# Polyfit
# ---
p1 = np.polyfit(x_train, y, 1)
R1 = np.mean((y-np.polyval(p1,x_train))**2)
ax.plot(x, np.polyval(p1,x), 'g-', linewidth=3, label="$\\hat{y}$ at order 1 with $\\hat{R}(f,d)=$ %.2f"%(R1))

p2 = np.polyfit(x_train, y, 2)
R2 = np.mean((y-np.polyval(p2,x_train))**2)
ax.plot(x, np.polyval(p2,x), 'r-', linewidth=3, label="$\\hat{y}$ at order 2 with $\\hat{R}(f,d)=$ %.2f"%(R2))

p3 = np.polyfit(x_train, y, 3)
R3 = np.mean((y-np.polyval(p3,x_train))**2)
ax.plot(x, np.polyval(p3,x), 'm-', linewidth=3, label="$\\hat{y}$ at order 3 with $\\hat{R}(f,d)=$ %.2f"%(R3))

p10 = np.polyfit(x_train, y, 10)
R10 = np.mean((y-np.polyval(p10,x_train))**2)
ax.plot(x, np.polyval(p10,x), '-', linewidth=3, label="$\\hat{y}$ at order 10 with $\\hat{R}(f,d)=$ %.2f"%(R10))


ax.grid(True)
for tick in ax.get_xticklabels():
    tick.set_fontsize(15)
for tick in ax.get_yticklabels():
    tick.set_fontsize(15)

ax.set_xlabel("$x$", fontsize=15)
ax.legend(loc='best', fontsize=15)
ax.set_xlim(-5,5)
ax.set_ylim(-80,20)

fig.set_size_inches(10,8)

plt.show()
```

<font size="3">$\hat{R}(f,d)$ is an estimate of the empirical risk (or training error). Most ML algorithms implement this empirical risk minimization. See how this risk is minimized as the polynomial degree is increased. However, we can clearly see that the blue curve does not generalize on the true data (i.e., the black curve). This simple example is a good observation of the underfitting/overfitting issue. 
    

The aim of a neural network is to learn how to map the input onto the outputs (i.e., learn mapping functions). The [**capacity**](https://machinelearningmastery.com/how-to-control-neural-network-model-capacity-with-nodes-and-layers/) of a network corresponds to the range of the types of mapping functions that the model can approximate/learn. The capacity is controlled by a set of hyper-parameters of the learning algorithm. For a polynomial fitting, the capacity of the model is directly linked to the interpolant degree. For a neural network, these hyper-parameters can be: the number of layers, the number of neurons per layer, the number of training iterations, and regularization terms, ... If the capacity is too small, the model underfits the data (e.g. green curve). If the capacity is too high, the model overfits the data (e.g. blue curve). 

Hence, to avoid this issue, an estimate of the risk can be computed on a distinct set, called the **validation set** to evaluate the performance of the model and to refine the hyperparameters (to avoid overfitting). It should not be used for model selection. 
</font>

### Datasets and `DataLoader`

<font size="3">There exists online databases that you can download and use as it is. We will do that in the practice section. However, most of the time, you have our own data and you want to make them accessible for the training. `PyTorch` offers the `DataLoader` class. For a complete review of `Data`, we encourage you to read [this](https://pytorch.org/docs/stable/data.html). Before using a `DataLoader`, we need to define a dataset. It exists two type of datasets: (a) the map-style datasets, and (b) iterable-style datasets. According to the documentation, 

* *A map-style dataset is one that implements the `__getitem__()` and `__len__()` protocols, and represents a map from (possibly non-integral) indices/keys to data samples.*

* *An iterable-style dataset is an instance of a subclass of IterableDataset that implements the `__iter__()` protocol, and represents an iterable over data samples. This type of datasets is particularly suitable for cases where random reads are expensive or even improbable, and where the batch size depends on the fetched data.*

Let us construct an iterable-style dataset for the fitting of our quadratic function. 

<div class="alert alert-block alert-warning">
<b>Info:</b> The library $\texttt{sklearn}$ also offers the opportunity to create database and to split them into training and validation database, as well as splitting according to K-fold cross-validation.
</div>
 </font>

```python
class MyIterableDataset(torch.utils.data.IterableDataset):
    def __init__(self, x, perc, epsilon, train=True):
        super(MyIterableDataset).__init__()
        
        self.nbData = x.shape[0]
        self.x      = x
        self.y      = fun(self.x).view((self.nbData,1)).to(torch.float32)
        #self.y     += torch.from_numpy(epsilon).view((self.nbData,1)).to(torch.float32)
        
        self.train  = train
        self.test   = not(train)
        self.size   = int(self.nbData*(1-perc))*self.test + int(self.nbData*perc)*self.train
        self.offset = int(self.nbData*(1-perc))*self.train
        
    def __iter__(self):
        worker_info = torch.utils.data.get_worker_info()
        if worker_info is None:  # single-process data loading, return the full iterator
            iter_start = self.offset
            iter_end   = self.offset + self.size
        else:  
            print("not implemented, read the documentation")
        index = torch.arange(iter_start,iter_end)
        buf   = torch.cat((self.x[index,:], self.y[index,:]), 1)
        return iter(buf)
```

<font size="3"> 
<div class="alert alert-block alert-info">
<b>Remark:</b> We still not discuss the normalization of the database. This step is crucial. First, you may non-dimensionalize your data according to the problem you treat (e.g., using a velocity scale, a length scale, a time scale, ...). Secondly, you further need to transform/scale your data to train the model. You may scale your data between 0 and 1. This scaling uses the minimum and maximum of your training set. You can also normalize your data by using the mean and variance of the training set. 
$$y^\ast = \dfrac{y - \min_y}{\max_y - \min_y} \hspace{5mm} \mbox{ or } \hspace{5mm} y^\star = \dfrac{y -\mu_y}{\sigma_y}$$
</div>
</font>

```python
nbData = 2000
x      = torch.linspace(-2,2,nbData).view((nbData,1))
rand_indx = torch.randperm(len(x))

batch_size  = 32
epsilon     = np.random.normal(0,0.5,x.shape[0])*0
trainData   = MyIterableDataset(x[rand_indx], 0.9, epsilon, train=True)
trainLoader = torch.utils.data.DataLoader(trainData, batch_size=batch_size, num_workers=0, pin_memory=False)

testData   = MyIterableDataset(x[rand_indx], 0.9, epsilon, train=False)
testLoader = torch.utils.data.DataLoader(testData, batch_size=batch_size, num_workers=0)
```

```python
# computation of the test loss
def test_loss(testLoader):
    test_loss = 0
    nb        = 0
    for data in testLoader:
        batch_sz = data.shape[0]
        x        = data[:,0].view((batch_sz,1))
        y        = data[:,1].view((batch_sz,1))
        y_pred   = mlp(x)        
        batch_loss = ((y - y_pred)**2).mean()
        test_loss += batch_loss.detach()
        nb         += 1
        
    return test_loss/nb

# model
mlp = MySequentialMLP(in_size,hidden_units,out_size)

# creation of an instance of the SGD class
#sgd_optimizer = optim.SGD(params=mlp.parameters(), lr=.001)
adam_optimizer = torch.optim.Adam(mlp.parameters(), lr=0.01, 
                                  betas=(0.9, 0.999), eps=1e-08, 
                                  weight_decay=0, amsgrad=False)

# learing steps
nbEpochs = 1000

loss     = np.zeros(nbEpochs)
lossTest = np.zeros(nbEpochs)

myTimer = time.time()
print("Beginning of the training ...")
for i in range(nbEpochs):
    mlp.train()

    train_loss = 0
    nb         = 0
    for data in trainLoader:
        batch_sz = data.shape[0]
        x        = data[:,0].view((batch_sz,1))
        y        = data[:,1].view((batch_sz,1))
        y_pred   = mlp(x)

        #  set all the grad values of the parameters of our net to zero
        adam_optimizer.zero_grad()        
        
        # computation of the loss and its gradient
        batch_loss = ((y - y_pred)**2).mean()
        batch_loss.backward()
    
        # update of the parameters 
        adam_optimizer.step()
        
        # save loss
        train_loss += batch_loss.detach()
        nb         += 1
    
    # always record your loss
    loss[i]     = train_loss/nb
    lossTest[i] = test_loss(testLoader)
    if(i%10==0):
        print("Epoch: %6d, train_loss=%.6e, val_loss=%.6e" %(i,loss[i],lossTest[i]))
print("... end of the training after %.3f sec" %(time.time() - myTimer))
```

<b>Remark 2:</b> One can also play with the batch size. First of all, there is the memory aspect behind the batch notion. Indeed, if you deal with a very huge database, loading it in one shot is impossible. Hence, the database is subdivided into small batches that fit in memory. Moreover, the batch size is seen as one of the most important hyperparameters to tune. It is known that a too-large batch size leads to a poor generalization of the model. Using a small batch size has faster convergence to "good" solutions (empirically shown!). Note that with a small batch size, it is not guaranteed that the model will converge to the global minima (in the case of convex functions).
